### 3.3 Data cleaning <a class="anchor" id="section303"></a>

```python
# List of columns that are supposed to contain numeric data
cols2numeric = ['car','price','body','mileage', 'engV','engType','registration','year','model','drive']

print("***** Count of Non Numeric Elements - Columnwise ******")
for col in cols2numeric:
    if(pd.to_numeric(carSales_data[col],errors='coerce').isnull().sum() > 0):
        print(col + ": " + str(pd.to_numeric(carSales_data[col],errors='coerce').isnull().sum()))
```

#### cleaning columns price and mileage

```python
# cleaning column Price where car Price = 0

carSales_data.price[carSales_data.price ==0].count()
```

#### Replace 0 value of price by median value

```python
# Replace 0 value of price by median walue-- because in order to use mean we need to remove the outliers otherwise mean will be affected

carSales_data['price']= carSales_data['price'].replace(0,carSales_data['price'].median())
carSales_data 
```

```python
carSales_data.mileage[carSales_data.mileage ==0].count()
```

#### Drop rows where mileage = 0 

```python
# cleaning column mileage - drop rows where mileage = 0

carSales_data = carSales_data.drop(carSales_data[carSales_data.mileage == 0].index)
```

```python
carSales_data.mileage[carSales_data.mileage ==0].count()
```

#### Removing outliers in mileage, mileage range= [0,50]. Mileage >60 is an outlier: replace it with NaN

```python
# removing the outliers in mileage-- 0 to 0

def clean_mileage(mileage):                                                     # Create a new column "mileage_clean" that replaces out-of-range mileages with "NaN"
    if mileage<=60:                                             
        return mileage
    else:
        return np.nan
carSales_data['mileage_clean'] = carSales_data['mileage'].apply(clean_mileage)

                                                          # Check out the new column and make sure it looks right
print("Minimum value: ", carSales_data["mileage_clean"].min())
print("Maximum value: ", carSales_data["mileage_clean"].max())
print("How many values are NaN?: ", pd.isnull(carSales_data['mileage_clean']).sum())
print("Frequency table for mileage_clean:")
carSales_data["mileage_clean"].value_counts().sort_index(0)
#carSales_data_clean = carSales_data_clean.drop(carSales_data_clean[carSales_data_clean.mileage > 60].index)
#carSales_data_clean
```

#### Count NaN values in price, mileage and engV

```python
PriceNaNCount = len(carSales_data) - carSales_data['price'].count()
MileageNaNCount = len(carSales_data) - carSales_data['mileage'].count()
engVNaNCount = len(carSales_data) - carSales_data['engV'].count()

print("Number of NaN values in Price - ", PriceNaNCount)
print("Number of NaN values in mileage - ", MileageNaNCount)
print("Number of NaN values in engV - ", engVNaNCount)
```

#### Replace NaN value for column engV by Mode

```python
# Replace engV NaN value by Mode

carSales_data['engV'].mode()
```

```python
#carSales_data['engV']= carSales_data['engV'].replace('NaN',carSales_data['engV'].mode())

carSales_data['engV'] = carSales_data['engV'].fillna(carSales_data['engV'].mode()[0])
```

## 4.17 Frequency distribution in a nutshell for Engine Type, Body Type, Registration done and Drive Type (Univariate Analysis)<a class="anchor" id="section417"></a>

```python
plt.figure(1)
plt.subplot(221)
carSales_data['engType'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='red')
plt.title("Engine Type frequency diagram")
plt.ylabel('Frequency of Engine Type')
plt.xlabel('Type of Engine');


plt.subplot(222)
carSales_data['body'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='green')
plt.title("Body Type frequency diagram")
plt.ylabel('Frequency of Body Type')
plt.xlabel('Type of Body');

plt.subplot(223)
carSales_data['registration'].value_counts(normalize= True).plot(figsize=(10,8),kind='bar',color='purple')
plt.title("Number of Registrations frequency diagram")
plt.ylabel('Frequency of registration')
plt.xlabel('Registration done?');

plt.subplot(224)
carSales_data['drive'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='orange')
plt.title("Drive Type frequency diagram")
plt.ylabel('Frequency of drive type')
plt.xlabel('Type of drive');
plt.tight_layout()
plt.show()
```

Findings-

1. More than 45 % of the cars sold has 'Petrol' type of Engine
2. Sedan constitutes more than 35% of the total sales
3. Less than 10% of cars are not registered
4. Most produced vehicle are of drive type 'front' around 60% followed by 'full' drive 25%
